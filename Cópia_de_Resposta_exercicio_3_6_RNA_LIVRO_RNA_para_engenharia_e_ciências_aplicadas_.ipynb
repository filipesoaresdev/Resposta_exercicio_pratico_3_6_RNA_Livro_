```
import numpy as np
import pandas

class Perceptron(object):

    def __init__(self, no_of_inputs, threshold=1000, learning_rate=0.01):
        self.threshold = threshold
        self.learning_rate = learning_rate
        self.weights = np.random.rand(no_of_inputs+1) 
           
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation > 0:
          activation = 1
        else:
          activation = -1            
        return activation

    def train(self, training_inputs, labels):
        print("Pesos iniciais : "+ str(self.weights))
        count = 0;
        possuiErro = True
        while possuiErro:
            count+=1
            possuiErro = False
           
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                erro = label - prediction
                if erro != 0:
                    possuiErro = True
                self.weights[1:] += self.learning_rate * erro * inputs
                self.weights[0] += self.learning_rate * erro
        print("Pesos finais : "+ str(self.weights))        
        print("Quantidade de épocas : "+ str(count))

class Predictions(object):
    def __init__(self, perceptron):
        self.perceptron = perceptron
    def realizarPredictions(self):
        inputs = np.array([-0.3665,0.0620,5.9891])
        print( "-0.3665,0.0620,5.9891 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([-0.7842,1.1267,5.5912])
        print( "-0.7842,1.1267,5.5912 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([0.3012,0.5611,5.8234])
        print( "0.3012,0.5611,5.8234 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([0.7757,1.0648,8.0677])
        print( "0.7757,1.0648,8.0677 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([0.1570,0.8028,6.3040])
        print( "0.1570,0.8028,6.3040 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([-0.7014,1.0316,3.6005])
        print( "-0.7014,1.0316,3.6005 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([0.3748,0.1536,6.1537])
        print( "0.3748,0.1536,6.1537 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([-0.6920,	0.9404,4.4058])
        print( "-0.6920,0.9404,4.4058 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([-1.3970,0.7141,4.9263])
        print( "-1.3970,0.7141,4.9263 ->>" +str(self.perceptron.predict(inputs)))
        inputs = np.array([-1.8842,-0.2805,1.2548])
        print( "-1.8842,-0.2805,1.2548 ->>" +str(self.perceptron.predict(inputs)))


input_dim = 3


learning_rate = 0.01


Training_Data = pandas.read_excel("/content/Tabela#Seção3.6_RNA.xls")



Expected_Output = Training_Data.d


Training_Data = Training_Data.drop(['d'], axis=1)

# T1
print("---- T1 -----")
perceptron1 = Perceptron(3)
perceptron1.train(Training_Data.values,Expected_Output.values)
predictions = Predictions(perceptron1)
predictions.realizarPredictions()


# T2
print("---- T2 -----")
perceptron2 = Perceptron(3)
perceptron2.train(Training_Data.values,Expected_Output.values)
predictions = Predictions(perceptron2)
predictions.realizarPredictions()

# T3
print("---- T3 -----")
perceptron3 = Perceptron(3)
perceptron3.train(Training_Data.values,Expected_Output.values)
predictions = Predictions(perceptron3)
predictions.realizarPredictions()

# T4
print("---- T4 -----")
perceptron4 = Perceptron(3)
perceptron4.train(Training_Data.values,Expected_Output.values)
predictions = Predictions(perceptron4)
predictions.realizarPredictions()

# T5
print("---- T5 -----")
perceptron5 = Perceptron(3)
perceptron5.train(Training_Data.values,Expected_Output.values)
predictions = Predictions(perceptron5)
predictions.realizarPredictions()


```
